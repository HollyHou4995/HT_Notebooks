```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
current_path = os.getcwd()
print(current_path)
```

```python
df = pd.read_csv('Advisory.csv')
df.head(5)
```

```python

# Assuming 'df' is your DataFrame and it contains a column named 'Title'
title_count = df['Title'].value_counts()

# Creating the plot
plt.figure(figsize=(10, 6))
ax = title_count.plot(kind='bar')  # 'ax' is the AxesSubplot object
plt.xlabel('Title Categories')
plt.ylabel('Count')
plt.title('Counts of Titles by Category')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
# Adding labels to each bar
ax.bar_label(ax.containers[0])

plt.show()  # This will display the plot
```

```python
attend_df = df[df['Status'] == 'Accepted']
attend_df.info()
```

```python
regi_count = attend_df['Regis_type'].value_counts()
regi_count
```

```python
# Data for the pie chart
labels = ['Staff', 'Member','Speaker']
sizes = [44.44, 44.44, 11.12 ]  # Proportions of each category
colors = ['lightcoral', 'lightskyblue', 'yellow']
explode = (0.1, 0, 0)  # "explode" the 1st slice (i.e., 'Category A')

# Plotting the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Registration Types of 2024 Advisory Board Attendees')
plt.axis('equal')

# Display the chart
plt.show()
```
